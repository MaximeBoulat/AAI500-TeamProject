```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Datasets/Credit.csv')
df = df.drop(columns=["ID"])

# Create AGE groups
age_bins = [20, 30, 40, 50, 60, 70, 80]
age_labels = ['20s', '30s', '40s', '50s', '60s', '70s']
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=age_bins, labels=age_labels, right=False)

# Define population segments
segment_columns = ['SEX', 'EDUCATION', 'MARRIAGE', 'AGE_GROUP']
segment_group = df.groupby(segment_columns)

# Count total and on-time payments per segment
segment_stats = segment_group['default payment next month'].agg(
    total='count',
    on_time=lambda x: (x == 0).sum()
).reset_index()

# Calculate P(On-time | Segment)
segment_stats['P_on_time_given_segment'] = segment_stats['on_time'] / segment_stats['total']

# Sort segments by highest probability
segment_stats_sorted = segment_stats.sort_values(by='P_on_time_given_segment', ascending=False)

# Create readable segment labels
segment_stats_sorted['Segment'] = (
    "SEX=" + segment_stats_sorted['SEX'].astype(str) +
    ", EDU=" + segment_stats_sorted['EDUCATION'].astype(str) +
    ", MARR=" + segment_stats_sorted['MARRIAGE'].astype(str) +
    ", AGE=" + segment_stats_sorted['AGE_GROUP'].astype(str)
)

# Plot top 10 segments
plt.figure(figsize=(12, 6))
sns.barplot(data=segment_stats_sorted.head(10), 
            x='P_on_time_given_segment', 
            y='Segment', 
            palette='viridis')
plt.xlabel("P(On-time Payment | Segment)")
plt.ylabel("Population Segment")
plt.title("Top 10 Segments Most Likely to Pay On Time")
plt.tight_layout()
plt.show()


# Printing my explanation
print("Explanation -- Bayesâ€™ Theorem to predict which population segment is most likely to pay their credit card bills on time")
print()
print("Calculates likelihood of on-time payment per segment using a Bayesian logic.")
print("Displays the top 10 most reliable segments to pay there credit on time.") 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Datasets/Credit.csv')
df = df.drop(columns=["ID"])

# Create AGE groups
age_bins = [20, 30, 40, 50, 60, 70, 80]
age_labels = ['20s', '30s', '40s', '50s', '60s', '70s']
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=age_bins, labels=age_labels, right=False)

# Define population segments
segment_columns = ['SEX', 'EDUCATION', 'MARRIAGE', 'AGE_GROUP']
segment_group = df.groupby(segment_columns)

# Count total and on-time payments per segment
segment_stats = segment_group['default payment next month'].agg(
    total='count',
    on_time=lambda x: (x == 0).sum()
).reset_index()

# Calculate P(On-time Segment)
segment_stats['P_on_time_given_segment'] = segment_stats['on_time'] / segment_stats['total']

# Plot histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(segment_stats['P_on_time_given_segment'], bins=20, kde=True, color='skyblue', edgecolor='black')

# Add vertical lines
mean_prob = segment_stats['P_on_time_given_segment'].mean()
max_prob = segment_stats['P_on_time_given_segment'].max()

plt.axvline(mean_prob, color='orange', linestyle='--', linewidth=2, label=f'Mean: {mean_prob:.2f}')
plt.axvline(max_prob, color='green', linestyle='-', linewidth=2, label=f'Max: {max_prob:.2f}')

# Labels and legend
plt.title("Distribution of P(On-time Payment Segment) with KDE")
plt.xlabel("P(On-time Payment Segment)")
plt.ylabel("Number of Segments")
plt.legend()
plt.tight_layout()
plt.show()

# Printing my explanation
print("Explanation: I analyzed the probability distribution of a client being granted a credit loan based on their credibility. While the dataset doesn't explicitly say if a client is -granted-- a loan, we can infer credibility based on the target variable")
print()
print("Plan:Define Credibility: Clients with default payment next month == 0. Analyze how credibility varies with key features (e.g., LIMIT_BAL, PAY_0, AGE, EDUCATION, etc.).Estimate probability:P(Credible Feature) Visualize the probability distribution using a plot (e.g., KDE + Histogram)")
print("This Chart Shows: The likelihood of being a credible client (on-time payer) based on credit limit. It looks like creditworthiness (credibility) shifts with the size of the loan") 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and prepare data
df = pd.read_csv('Datasets/Credit.csv')
df = df.drop(columns=["ID"])
df['Credible'] = df['default payment next month'].apply(lambda x: 1 if x == 0 else 0)

# Set plot style
sns.set(style="whitegrid")

# -PLOT : Credibility vs Education 
plt.figure(figsize=(8, 6))
edu_order = sorted(df['EDUCATION'].unique())  # Ensure consistent order
sns.barplot(data=df, x='EDUCATION', y='Credible', estimator=lambda x: x.mean(), order=edu_order, palette="viridis")
plt.title("Probability of Being Credible by Education Level")
plt.xlabel("Education Level (1=Graduate, 2=University, 3=High School, etc.)")
plt.ylabel("P(Credible | Education)")
plt.tight_layout()
plt.show()

print("Explanation -- additional plots that analyze the probability of a client being credible (on-time payer) by: Education level")
print()
print("The Plot shows how credibility varies across education levels. Example: Higher-educated individuals may show higher credibility rates.")
```
